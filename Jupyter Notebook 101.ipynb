<span style="font-family:Courier New; font-size:16px;  text-transform: capitalize;">
<b>Jupyter Notebook</b> is a web-based computing platform that allows developers to code, visualize, share and embed multimedia with explanatory text inside a single document. It is a famous tool for showcasing your work because You can see both the code and the results in the same file. 

It is so much among data scientists because you can run each line of code separately and generate results to get a better understanding. Jupyter Notebook allows its users to download the notebook in various file formats like PDF, HTML, Python, Markdown, or an. ipynb file. 

    > There Are Two Essential Parts of a Jupyter Notebook ➡ Markdown Cells and Code Cells. 

**1. Markdown Cell**<br>
A Markdown Cell represents your work by providing enough explanation to readers about the code block below them. Markdown Writing Is an essential skill for every data scientist. It displays text which is formatted using a markup language. By Default All The Cells Will Be Code Cells. To Convert a Code Cell Into a Markdown Cell Just Press `ESC` to enter in command mode and then press `m` to convert it into a markdown.

**2. Code Cell**<br>
A Code Cell allows you to Create, Update, Read, Edit, and Delete Code With full syntax highlighting and tab completion. The Default Kernal of Jupyter Notebook is Ipython That Runs Python Code. When a Code Cell has executed the code inside it is sent to the kernel (IPython default) that runs the code. To Execute a Code Cell You Can Use Keyboard Keys Combination `SHIFT+ENTER` 

***Installation*** <br>
To Install Jupyter Notebook Inside a python environment just run the command `pip install jupyter` and to open the web client-server just type `jupyter notebook`. After Running The Command A Web Page Will Open On Your Default Browser That is The Homepage of Jupyter Notebook. Simply click on New select python3 to create a notebook.

## 1.3 Creating Tables

<br>
<span style="font-family:Courier New; font-size:16px;  text-transform: capitalize;">
    
There Are Two Methods To Create Tables Inside Jupyter Notebook. First One is by using `|` as a table divide. The Second One is with the help of `<table> </table>` tag. You Can Use `:-:` for centered, `:-` for left centered and `—` for right centered table.
 
    
</span>

▶ **Syntax For `|`**

```
| SN | Column A | Column B | Column C |
| :---: | :---: | :---: | :---: | 
| 0   | 10 | 11 | 100 |
| 1 | 22.5 | 23 | 23.5 |

```

#### OUTPUT
